```python
import tensorflow as tf
IMG_SIZE = (224, 224)
BATCH = 32

# Folders se images ko load karna (Aapke folder structure ke mutabiq sahi path)
train_ds = tf.keras.utils.image_dataset_from_directory(
    'data/real_vs_fake/real-vs-fake/train',
    image_size=IMG_SIZE,
    batch_size=BATCH
)
val_ds = tf.keras.utils.image_dataset_from_directory(
    'data/real_vs_fake/real-vs-fake/valid',
    image_size=IMG_SIZE,
    batch_size=BATCH
)
test_ds = tf.keras.utils.image_dataset_from_directory(
    'data/real_vs_fake/real-vs-fake/test',
    image_size=IMG_SIZE,
    batch_size=BATCH
)

class_names = train_ds.class_names
print("Classes:", class_names)
print("Train batches:", len(train_ds))
print("Val batches:", len(val_ds))
print("Test batches:", len(test_ds))
```

```python
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.cache().shuffle(1000).prefetch(AUTOTUNE)
val_ds   = val_ds.cache().prefetch(AUTOTUNE)
test_ds  = test_ds.cache().prefetch(AUTOTUNE)
print("✅ Data pipeline ready! Member 2 ko handover karo.")
```
